```python
types_of_coustomers = df["Segment"].unique()
print(types_of_coustomers)
```

```python
number_of_customers = df["Segment"].value_counts().reset_index()
number_of_customers = number_of_customers.rename(columns={"Segment":"Coustomer Type", "count":"Total Coustomers"})
print(number_of_customers)
```

```python
plt.pie(number_of_customers['Total Coustomers'], labels=number_of_customers["Coustomer Type"], autopct = '%1.1f%%')
plt.title('Distribution Of Customers')
plt.show()
```

```python
#Sales per coustomer category
Sales_per_category = df.groupby('Segment')['Sales'].sum().reset_index()
Sales_per_category = Sales_per_category.rename(columns={'Segment':'Customer Type', 'Sales':'Total Sales'})
print(Sales_per_category)
```

```python
plt.pie(number_of_customers['Total Coustomers'], labels = number_of_customers['Coustomer Type'], autopct='%1.1f%%')
plt.title('Distribution of Coustomers')
plt.show()
```

```python
plt.bar(Sales_per_category['Customer Type'], Sales_per_category['Total Sales'])
plt.title('Sales per Customer')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')
plt.show()
```

```python
order_freq = df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Order ID'].count().reset_index()
order_freq = order_freq.rename(columns={'Order ID' :'Total Orders'})
repeat_order = order_freq[order_freq['Total Orders']>1]
sorted_orders = repeat_order.sort_values(by = 'Total Orders', ascending =False)
print(sorted_orders.head(10).reset_index(drop= True))
```

```python
customer_sales = df.groupby(['Customer ID', 'Customer Name','Segment'])['Sales'].sum().reset_index()
top_spenders = customer_sales.sort_values(by ='Sales', ascending = False)
print(top_spenders.head(10).reset_index(drop= True))
```

```python
shipping_class = df['Ship Mode'].value_counts().reset_index()
shipping_class = shipping_class.rename(columns={'Ship Mode' : 'Mode of Transport','count':'Frequency'})
print(shipping_class)
```

```python
plt.pie(shipping_class['Frequency'],labels = shipping_class['Mode of Transport'], autopct = '%1.1f%%')
plt.title('Popular Shipping Choices')
plt.show()
    
```

```python
state = df['State'].value_counts().reset_index()
state = state.rename(columns={'State':'State of Delivery','count':'Frequency of deliveries'})
print(state)
```

```python
city = df['City'].value_counts().reset_index()
city= city.rename(columns ={'City': 'City of Deliveries', 'count':'Frequencies of deliveries'})
print(city.head(10).reset_index(drop= True))
```

```python
state_sales = df.groupby(['State'])['Sales'].sum().reset_index()
top_state_sales = state_sales.sort_values(by ='Sales', ascending = False)
print(top_state_sales.head(30).reset_index(drop=True))
```

```python
city_sales = df.groupby(['City'])['Sales'].sum().reset_index()
greater_sales = city_sales[city_sales["Sales"]> 10000]
top_city_sales = greater_sales.sort_values(by='Sales', ascending = False)
print(top_city_sales.reset_index(drop = True))
```

```python
subcategory_count = df.groupby(['Category'])['Sub-Category'].nunique().reset_index()
subcategory_count = subcategory_count.sort_values(by = 'Sub-Category', ascending = False)
print(subcategory_count.reset_index(drop=True))
```

```python
plt.pie(category_sales['Sales'], labels = category_sales['Category'], autopct = '%1.1f%%')
plt.title('Distribution of Sales among Categories')
plt.show()
```

```python
sub_category_sales = df.groupby(['Sub-Category'])['Sales'].sum().reset_index()
sub_category_sales = sub_category_sales.sort_values(by = 'Sales', ascending = True)
print(sub_category_sales.reset_index(drop=True))
```

```python
plt.bar(sub_category_sales['Sub-Category'], sub_category_sales['Sales'])
plt.title("Top Product Subcategories according to their sales")
plt.xlabel('Product Sub-Categories')
plt.ylabel('Sales')
plt.show()
```

```python
df['Order Date']= pd.to_datetime(df['Order Date'], dayfirst = True)
yearly_sales =  df.groupby(df['Order Date'].dt.year)['Sales'].sum()
yearly_sales = yearly_sales.reset_index()
yearly_sales = yearly_sales.rename(columns={'Order Date' : 'Year','Sales':'Annual Sales'})
print(yearly_sales)
```

```python
plt.bar(yearly_sales['Year'], yearly_sales['Annual Sales'])
plt.title("Sales of The Company along the years")
plt.xlabel('Year')
plt.ylabel('Annual Sales')
plt.show()
```

```python
plt.plot(yearly_sales['Year'], yearly_sales['Annual Sales'], marker='o', linestyle='--')
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)
year_sales = df[df['Order Date'].dt.year == 2015]
quaterly_sales = year_sales.resample('QE', on = 'Order Date')['Sales'].sum()
quaterly_sales = quaterly_sales.reset_index()
quaterly_sales = quaterly_sales.rename(columns = {'Order Date' :'Quater', 'Sales':'Quaterly Sales'})
print(quaterly_sales)
```

```python
plt.plot(quaterly_sales['Quater'], quaterly_sales['Quaterly Sales'], marker = 'o', linestyle ='--')
plt.title('Quaterly Sales For the year 2018')
plt.xlabel('Quater')
plt.ylabel('Sales')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)
yearly_sales = df[df['Order Date'].dt.year == 2018]
monthly_sales = yearly_sales.resample('ME', on='Order Date')['Sales'].sum()
monthly_sales = monthly_sales.reset_index()
monthly_sales = monthly_sales.rename(columns={'Order Date':'Month','Sales':'Monthly Sales'})
print(monthly_sales)
```

```python
plt.plot(monthly_sales['Month'], monthly_sales['Monthly Sales'], marker = 'o', linestyle='--')
plt.title('Monthly Sales For the year 2018')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
